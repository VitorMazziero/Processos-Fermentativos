#1. Contínuo sem Reciclo

Contínuo.png

Um processo contínuo sem reciclo é carecterizado pela entrada e saída de meio, de maneira que o volume presente no reator não varie, sem realimentação de células no processo.

Portanto, no estado estacionário a variação mássica que representa o acúmulo é zero:


>$\frac{\mathrm{dm} }{\mathrm{d} t} = 0$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀Eq. (1)

##1.1. Balanço de Massa Geral

Para Contínuo, não há variação de volume, portanto, a vazão volumétrica de entrada deve ser igual a de saída, como apresentado pela Eq.(2):

>$Q_{e} = Q$  ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (2)

##1.2. Balanço de Massa Celular

Acúmulo = Entra - Sai + Geração - Consumo

⠀

Para Contínuo o balanço geral de células é simplificado para:


> 0 $=$ Entra - Sai + Geração - Morte

⠀⠀⠀⠀⠀⠀⠀⠀⠀$\therefore$

> $0= F_{xe} - F_{x} + R_{x} - R_{d}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (3)



---

Analogamente ao semicontínuo, define-se a **taxa de diluição** como $D = \frac{Q_{e}}{V} = \frac{Q}{V} $


---



Substituindo massa $m_{x}$ pela concentração celular $C_{x}$ utilizando a relação $C_{x}=\frac{m_{x}}{V}$, a vazão mássica $F_{xe}$ passa a ser descrita por $Q_{e}C_{xe}(\frac{m^3}{h}\frac{kg}{m^3})$, a geração celular passa a ser descrita por $r_{x}=\frac{R_{x}}{V}$ e a morte celular por $r_{d}=\frac{R_{d}}{V}$, dividindo por $V$ e reorganizando a equação em função de $D$:

$0 = Q_{e}C_{xe} - QC_{x} + R_{x} - R_{d}$
⠀

$\frac{0}{V} = \frac{QC_{xe}}{V} - \frac{QC_{x}}{V} + \frac{R_{x} - R_{d}}{V}$ ⠀

⠀⠀

>$D(C_{x} - C_{xe}) = r_{x} - r_{d}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (4)⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀


Analogamente a Batelada Simples e Alimentada, as taxas de crescimento e morte celular são descritas a seguir uitlizando a relação para reações de primeira ordem:

$r_{x}=\frac{\mathrm{d}[C] }{\mathrm{d} t} = k_{1}[C]^{1}$

$r_{d}=-\frac{\mathrm{d}[C] }{\mathrm{d} t} = k_{2}[C]^{1}$

---

Na qual: $[C]^{1} = C_{x}$ ; $k_{1}=\mu$ (taxa de crescimento celular) e $k_{2}=k_{d}$ (taxa de morte celular)

---

Então, a relação dada pela Eq.(4) passa a ser descrita pela Eq.(5) e a taxa de diluição é representada pela Eq.(6):

$D(C_{x} - C_{xe}) = (\mu - k_{d})C_{x}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (5)

⠀⠀

>$D = \frac{(\mu - k_{d})C_{x}}{C_{x} - C_{xe}}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (6)

No sistema em contínuo sem reciclo de células, considera-se a entrada de células $C_{xe} = 0$ e, para  $k_{d}<<\mu$, $D$ é representado pela Eq.(7)

$D = \frac{(\mu - k_{d})C_{x}}{C_{x}}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀

>$D = (\mu - k_{d})$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀Eq. (7)

##1.3. Balanço de Massa de Substrato

Acúmulo = Entra - Sai + Geração - Consumo

⠀

Para Contínuo o balanço geral de células é simplificado para:


> 0 $=$ Entra $-$ Sai $-$ Consumo

⠀⠀⠀⠀⠀⠀⠀⠀⠀$\therefore$

> $0= F_{se} - F_{s} - R_{s}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (8)

Utilizando a as relações de consumo de substrato $r_{s}=\frac{R_{s}}{V}$, de concentração de substrato $C_{s}=\frac{m_{s}}{V}$,  a vazão mássica $F_{s}$ passa a ser descrita por $QC_{s}(\frac{m^3}{h}\frac{kg}{m^3})$ e de taxa de diluição $D = \frac{Q}{V}$ reorganiza-se a Eq.(8) na Eq.(9):

0 = $Q_{e}C_{se} - QC_{s} -  R_{s} $

$\frac{Q_{e}C_{se}}{V} = \frac{QC_{s}}{V} + \frac{R_{s}}{V}$

$DC_{se}  = DC_{s} +  r_{s} $

⠀⠀

>$D(C_{se}-C_{s})  = r_{s} $⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (9)

Para relacionar o consumo de substrato com a formação celular, é necessário definir o **rendimento de crescimento celular** $Y_{\frac{x}{s}}$ (a quantidade de célula que é formada por substrato consumido):


$Y_{\frac{x}{s}} = \frac{r_{x}}{r_{s}}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀

⠀⠀

>$r_{s} = \frac{r_{x}}{Y_{\frac{x}{s}}}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀Eq. (10)

Unindo-se a Eq. (9) e (10) e utilizando a relação de $r_{x}=(\mu - k_{d})C_{x}$, a taxa de diluição $D$ é representada pela Eq.(11) quando $k_{d}<<\mu$ :

$D(C_{se}-C_{s})  = \frac{\mu C_{x}}{Y_{\frac{x}{s}}} $

⠀⠀
>$D  = \frac{(\mu - k_{d})C_{x}}{Y_{\frac{x}{s}}(C_{se}-C_{s})} $⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (11)



##1.4. Balanço de Massa de Produto

Acúmulo = Entra - Sai + Geração - Consumo

⠀

Para Contínuo o balanço geral de produto é simplificado para:


> 0 $=$ Entra - Sai + Gerado - Degradado

⠀⠀⠀⠀⠀⠀⠀⠀⠀$\therefore$

> $0 = F_{pe} - F_{p} + R_{p} - R_{dp}$⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (13)

Utilizando a as relações de consumo de substrato $r_{p}=\frac{R_{p}}{V}$, de concentração de produto $C_{p}=\frac{m_{p}}{V}$ e de taxa de diluição $D = \frac{Q}{V}$ reorganiza-se a Eq.(13) na Eq.(14):

$\frac{Qe}{V}C_{pe} = \frac{Q}{V}C_{p} + \frac{R_{dp}}{V} - \frac{R_{p}}{V}$


$D(C_{pe}-C_{p})  =  r_{dp} - r_{p}$

⠀⠀

>$D = \frac{r_{dp} - r_{p}}{C_{pe}-C_{p}} $⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (14)

Para relacionar a geração de produto com a formação celular, é necessário definir seu tipo de geração, que afetará o valor da concentração de produto obtida a partir da taxa de diluição da Eq.(14), na qual $\alpha$ e $\beta$ são constantes e $r_{dp} = k_{dp}C_{p}$:
⠀

⠀

>**Associado ao crescimento:** ⠀$D = \frac{k_{dp}C_{p} - \alpha \mu C_{x}}{C_{pe}-C_{p}} $

>$C_{p} = \frac{DC_{pe} + \alpha \mu C_{x}}{D+k_{dp}}$  ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ ⠀Eq. (15)

⠀

>**Não associado ao crescimento:**  $D = \frac{k_{dp}C_{p} - \beta C_{x}}{C_{pe}-C_{p}} $

>$C_{p} = \frac{DC_{pe} + \beta C_{x}}{D+k_{dp}}$  ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ ⠀Eq. (16)

⠀

>**Parcialmente associado ao crescimento:** ⠀$D = \frac{k_{dp}C_{p} - \alpha \mu C_{x}-\beta C_{x}}{C_{pe}-C_{p}} $

>$C_{p} = \frac{DC_{pe} + \alpha \mu C_{x} + \beta C_{x}}{D+k_{dp}}$  ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀Eq. (17)



## 1.5. Modelo cinético não estruturado para simulação **sem inibição**

Utilizando a Eq.(7) como base para $C_{s}$, Eq.(11) como base para $C_{x}$ e Eq.(17) para $C_{p}$, substitui-se $\mu$ constante por modelos cinéticos de $\mu$ variável. Para o modo contínuo vamos utilizar o modelo proposto por Monod (1949), entretanto quaisquer modelos já apresentados nos modos de operação de Batelada Simples e Alimentada podem ser substituidos.

### 1.5.1. (Monod, 1949)

$\mu = \frac{\mu_{max}C_{s} }{K_{s}+C_{s}}$

Utilizando o balanço celular para $C_{s}$:

$D = (\frac{\mu_{max}C_{s} }{K_{s}+C_{s}} - k_{d})$

Colocando $C_{s}$ em evidência:

>$ C_{s}= \frac{K_{s}D}{(\mu_{max}-k_{d}-D)}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (18)

Utilizando o balanço de substrato para $C_{x}$:

$D  = \frac{(\mu - k_{d})C_{x}}{Y_{\frac{x}{s}}(C_{se}-C_{s})} $

$D  = \frac{DC_{x}}{Y_{\frac{x}{s}}(C_{se}-C_{s})} $

Colocando $C_{x}$ em evidência:

$ C_{x}= \frac{{Y_{\frac{x}{s}}(C_{se}-C_{s})D}}{(\frac{\mu_{max}C_{s} }{K_{s}+C_{s}}- k_{d})}$

$ C_{x}= \frac{{Y_{\frac{x}{s}}(C_{se}-C_{s})D}}{D}$

>$ C_{x}= {Y_{\frac{x}{s}}(C_{se}-C_{s})}$ ⠀⠀⠀⠀ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (19)

Assim, unindo-se as Eq.(18) e (19):

>$ C_{x}= {Y_{\frac{x}{s}}(C_{se}-\frac{K_{s}D}{(\mu_{max}-k_{d}-D)})}$ ⠀⠀⠀⠀ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (20)

##1.6. Taxa de diluição ótima ($D_{max}$)


Partindo da definição de **produtividade volumética**, dado a seguir, calcula-se a diluição na qual a produtividade é máxima, ou seja, sua derivada é zero:

$Q_{x} = \frac{Q_{e}C_{x}}{V} = DC_{x}$

⠀⠀

---


Na qual $D=D_{max}$ quando $\frac{\mathrm{dQ_{x}} }{\mathrm{d} D} = 0$


---



$\frac{\mathrm{dDC_{x}} }{\mathrm{d} D} = 0$

$\frac{\mathrm{d}}{\mathrm{dD}}{Y_{\frac{x}{s}}(C_{se}-C_{s})}D = 0$

$\frac{\mathrm{d}}{\mathrm{dD}}{Y_{\frac{x}{s}}(C_{se}-\frac{K_{s}D}{(\mu_{max}-k_{d}-D)})}D = 0$

Derivando a equação anterior:


```python
import sympy as sp
from sympy import diff

D, Yxs, Cse, Ks, μmax, kd = sp.symbols('D Yxs Cse Ks μmax kd')
func=Yxs*(Cse-(Ks*D)/(μmax-kd-D))*D

diff(func,D)
```

Obtém-se o $D_{max}$ igualando a equação acima a zero e deixando $D$ em evidência
⠀
>$D_{max} = -\frac{\sqrt{(k_{d} - \mu_{max})^2 K_{s}(C_{se} + K_{s})}}{K_{s}+C_{se}}-k_{d}+\mu_{max}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (21)

---

Definindo a **tempo de residência** como $\tau = \frac{1}{D}$


---



Teoricamente, para a reação em contínuo entrar em estado estacionário, em geral demora 3x o tempo de residência

##1.7. Simulação da variação da diluição no estado estacionário



```python
# Simulação - Contínuo

# -*- coding: utf-8 -*-
"""
Created on Wed Sep  7 15:36:42 2022

@author: Vítor Teixeira Mazziero - UNESP Araraquara
"""

import warnings
import numpy as np
import matplotlib.pyplot as plt

# Concentrações de alimentação
Cxe=0
Cse=20.0
Cpe=0

# Parâmetros para crescimento celular
μmax = 0.5
kd = 0.05
Ks = 0.2

# Parâmetros para consumo de substrato
Yxs = 0.5

# Parâmetros para formação de produto
α = 0.1
β = 0.05
Kdp = 0

# Taxa de diluição (de 0 até 0.495)
D=np.arange(0.0001,0.58,0.0001) #[D]=1/h

# Concentração de Células (kg/m³)
Cs=(Ks*D)/(μmax-kd-D) #Eq.(18)

# Concentração de Células (kg/m³)
Cx=Yxs*(Cse-Cs) #Eq.(19)

# Concentração de Produto (kg/m³)
Cp=((D*Cpe)+(α*μmax*Cx)+(β*Cx))/(D+Kdp) #Eq.(17)

#Produtividade
P = Cx*D

# Dmax
Dmax = - np.sqrt((kd - μmax)**2 * Ks*(Cse + Ks))/(Cse + Ks) - kd + μmax

# Ignorar Concentrações Indefinidas
for i in range(len(D)):
    if Cs[i]>=Cse or Cs[i]<0 :
      Cs[i]=Cse
      Cp[i]=Cpe
      Cx[i]=Cxe
      P[i]=0

warnings.filterwarnings("ignore")

# Construção do Gráfico

plt.figure(figsize=(10,6), facecolor="white", edgecolor="black")

plt.plot(D, Cs, label = "Substrato", color="blue", linewidth=2, linestyle="-")
plt.plot(D, Cx, label = "Células", color="red", linewidth=2, linestyle="-")
plt.plot(D, Cp, label = "Produto", color="black", linewidth=2, linestyle="-")
plt.plot(D, P, label = "Produtiv", color="pink", linewidth=2, linestyle="--")

plt.xlabel("Taxa de Diluição (1/h)", fontdict={'fontsize':12})
plt.ylabel("Concentração (kg/m³)", fontdict={'fontsize':12})

plt.ylim([-5, 25])
plt.xlim([0, 0.55])

plt.axhline(0, color = 'r', linestyle = ':')
plt.axvline(Dmax, color = 'g', linestyle = ':')
plt.text(y=20, x=Dmax, s='Dmax = %.4f' %Dmax, fontsize = '12', position= (0.455,-4.7))

plt.show()
```

#2. Contínuo com Reciclo Externo

Contínuo_Reciclo_Externo.png

Um processo contínuo com reciclo externo é carecterizado pela entrada e saída constante de volume do processo, de maneira que o volume presente no reator não varie, com realimentação de células no processo. Essas células são concentradas em uma unidade de separação externa ao reator.

##2.1. Balanço de Massa Geral

Acúmulo = Entra - Sai + Geração - Consumo

⠀

Para Contínuo com reciclo externo o balanço geral com volume de controle no **reator**:


> Acúmulo $=$ Entra

⠀⠀⠀⠀⠀⠀⠀⠀⠀$\therefore$

$0= (F_{e}+F_{3}) - F_{1}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀
>$F_{1}= (F_{e}+F_{3})$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (22)

Para Contínuo com reciclo externo o balanço geral com volume de controle no **reator + separador**:


> Acúmulo $=$ Entra - Sai

⠀⠀⠀⠀⠀⠀⠀⠀⠀$\therefore$

$0= F_{1}-(F_{2} + F_{3})$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀
>$F_{1}= (F_{2}+F_{3})$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (23)

Unindo as Eq. (22) e (23):

>$(F_{e}+F_{3}) = (F_{2}+F_{3})$

⠀⠀⠀⠀⠀⠀⠀⠀⠀$\therefore$

⠀⠀⠀⠀⠀⠀$F_{e} = F_{2}$  ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (24)

---

Definindo a **razão de reciclo** como $r_{r} = \frac{F_{3}}{F_{2}}$


---



>$F{1}= F{e}(1+r_{r})$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (25)

##2.1. Balanço de Massa Celular

Acúmulo = Entra - Sai + Geração - Consumo

⠀

Para Contínuo com Reciclo externo o balanço geral de células é simplificado no estado estacionário:


>0 $=$ Entra - Sai + Geração - Morte

⠀⠀⠀⠀⠀⠀⠀⠀⠀$\therefore$

> $0= (F_{xe} + F_{x3}) - F_{x1} + R_{x} - R_{d}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (26)



---

Analogamente ao semicontínuo, define-se a **taxa de diluição** como $D = \frac{Q_{e}}{V}$ e utiliza-se as relações entre vazão mássica ($F$) e volumétrica ($Q$)


---



Utilizando as relações apresentadas nas Eq. (23) a (26):

$0= (\frac{Q_{e}}{V}C_{xe} + \frac{Q_{3}}{V}C_{x3}) - \frac{Q_{1}}{V}C_{x1} + r_{x} - r_{d}$

$0= (\frac{Q_{e}}{V}C_{xe} + \frac{Q_{3}}{V}C_{x3}) - \frac{ Q{e}(1+r_{r})}{V}C_{x1} + r_{x} - r_{d}$

$0= (DC_{xe} + {r_{r}}DC_{x3}) - { D(1+r_{r})}C_{x1} + r_{x} - r_{d}$



---

Definindo a **fator de reciclo celular** como $f_{rx} = \frac{C_{x3}}{C_{x1}}$, considerando a **mistura perfeita** no reator, tal que $C_{x1} = C_{x}$ e que **não há a entrada de células**, $C_{xe}=0$


---



$0= DC_{xe} + {r_{r}}Df_{rx}C_{x} - { D(1+r_{r})}C_{x} + (\mu - k_{d})C_{x}$

$0= C_{x}({r_{r}}Df_{rx} - {D(1+r_{r})} + (\mu - k_{d}))$


>$D = \frac{\mu - k_{d}}{1+r_{r}(1-f_{rx})}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (27)

##2.3. Balanço de Massa de Substrato

Acúmulo = Entra - Sai + Geração - Consumo

⠀

Para Contínuo o balanço geral de células é simplificado para:


> 0 $=$ Entra $-$ Sai $-$ Consumo

⠀⠀⠀⠀⠀⠀⠀⠀⠀$\therefore$

> $0= (F_{se}+F_{s3}) - F_{s1} - R_{s}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (28)

$0= (\frac{Q_{e}}{V}C_{se} + \frac{Q_{3}}{V}C_{s3}) - \frac{Q_{1}}{V}C_{s1} - r_{s}$

$0= (\frac{Q_{e}}{V}C_{se} + \frac{Q_{3}}{V}C_{s3}) - \frac{ Q{e}(1+r_{r})}{V}C_{s1} - r_{s}$

$0= (DC_{se} + {r_{r}}DC_{s3}) - { D(1+r_{r})}C_{s1} - r_{s}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀

Considerando apenas a separação das células no separador: $C_{s} = C_{s1} = C_{s2}= C_{s3}$

$0= DC_{se} + {r_{r}}DC_{s} - DC_{s}(1+r_{r}) + (\mu - k_{d})\frac{C_{x}}{Y_{\frac{x}{s}}}$

$0= D(C_{se} -C_{s}) - (\mu - k_{d})\frac{C_{x}}{Y_{\frac{x}{s}}}$


>$D =\frac{C_{x}(\mu - k_{d})}{Y_{\frac{x}{s}}(C_{se} -C_{s})}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (29)

##2.4. Balanço de Massa de Produto

Acúmulo = Entra - Sai + Geração - Consumo

⠀

Para Contínuo o balanço geral de células é simplificado para:


> 0 $=$ Entra $-$ Sai $+$ Gerado $-$ Consumo

⠀⠀⠀⠀⠀⠀⠀⠀⠀$\therefore$

> $0= (F_{pe}+F_{p3}) - F_{p1} + R_{p} -  R_{dp}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (30)

$0= (\frac{Q_{e}}{V}C_{pe} + \frac{Q_{3}}{V}C_{p3}) - \frac{Q_{1}}{V}C_{p1} + r_{p} - r_{dp}$

$0= (\frac{Q_{e}}{V}C_{pe} + \frac{Q_{3}}{V}C_{p3}) - \frac{ Q{e}(1+r_{r})}{V}C_{p1} + r_{p} - r_{dp}$

$0= (DC_{pe} + {r_{r}}DC_{p3}) - { D(1+r_{r})}C_{p1} + r_{p} - r_{dp}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (31)

Considerando apenas a separação das células no separador: $C_{p} = C_{p1} = C_{p2}= C_{p3}$ e $r_{dp} = k_{dp}C_{p}$, na qual $\alpha$ e $\beta$ são constantes:

$0= DC_{pe} + {r_{r}}DC_{p} - DC_{p}(1+r_{r}) + r_{p} - r_{dp}$

$0= D(C_{pe} -C_{p}) + r_{p} - r_{dp}$

Para relacionar a geração de produto com a formação celular, é necessário definir seu tipo de geração, que afetará o valor da concentração de produto obtida a partir da taxa de diluição da Eq.(14), na qual $\alpha$ e $\beta$ são constantes e $r_{dp} = k_{dp}C_{p}$:
⠀

⠀

>**Associado ao crescimento:** ⠀$D = \frac{k_{dp}C_{p} - \alpha \mu C_{x}}{C_{pe}-C_{p}} $

>$C_{p} = \frac{DC_{pe} + \alpha \mu C_{x}}{D+k_{dp}}$  ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ ⠀Eq. (32)

⠀

>**Não associado ao crescimento:**  $D = \frac{k_{dp}C_{p} - \beta C_{x}}{C_{pe}-C_{p}} $

>$C_{p} = \frac{DC_{pe} + \beta C_{x}}{D+k_{dp}}$  ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ ⠀Eq. (33)

⠀

>**Parcialmente associado ao crescimento:** ⠀$D =\frac{k_{dp}C_{p} - \alpha \mu C_{x} - \beta C_{x}}{C_{pe} -C_{p}}$

>$C_{p} = \frac{DC_{pe} + \alpha \mu C_{x} + \beta C_{x}}{D+k_{dp}}$  ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀Eq. (34)



## 2.5. Modelo cinético não estruturado para simulação **sem inibição**

Utilizando a Eq.(27) como base para $C_{s}$, Eq.(29) como base para $C_{x}$ e Eq.(34) para $C_{p}$, substitui-se $\mu$ constante por modelos cinéticos de $\mu$ variável. Para o modo contínuo vamos utilizar o modelo proposto por Monod (1949), assim como no contínuo sem reciclo.

### 2.5.1. (Monod, 1949)

$\mu = \frac{\mu_{max}C_{s} }{K_{s}+C_{s}}$

Utilizando o balanço celular para $C_{s}$:

$D = \frac{\frac{\mu_{max}C_{s} }{K_{s}+C_{s}} - k_{d}}{1+r_{r}(1-c)}$

Colocando $C_{s}$ em evidência:

>$ C_{s}= \frac{K_{s}[D(1+r_{r}(1-c))+k_{d}]}{\mu_{max}-k_{d}-D(1+r_{r}(1-c))}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (35)

Utilizando o balanço celular para $C_{x}$:

$D =\frac{C_{x}(\frac{\mu_{max}C_{s} }{K_{s}+C_{s}} - k_{d})}{Y_{\frac{x}{s}}(C_{se} -C_{s})}$

Colocando $C_{x}$ em evidência:


>$ C_{x}= \frac{{Y_{\frac{x}{s}}(C_{se}-C_{s})}D}{\frac{\mu_{max}C_{s} }{K_{s}+C_{s}} - k_{d}}$ ⠀⠀⠀⠀ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (36)

```python
# Simulação - Contínuo com Reciclo Externo

# -*- coding: utf-8 -*-
"""
Created on Wed Sep  7 15:36:42 2022

@author: Vítor Teixeira Mazziero - UNESP Araraquara
"""

import warnings
import numpy as np
import matplotlib.pyplot as plt

# Concentrações de alimentação
Cxe=0
Cse=20
Cpe=0

# Parâmetros para crescimento celular
μmax = 0.5
kd = 0
Ks = 0.1

# Parâmetros para consumo de substrato
Yxs = 0.5

# Parâmetros para formação de produto
α = 0.1
β = 0.05
Kdp = 0

# Taxa de diluição
D=np.arange(0.0001,1,0.0001) #[D]=1/h

#Constantes de reciclo
r=0.2 #[r]=Adimensional
c=3 #[c]=Adimensional

# -------------Processo sem reciclo-------------------

# Concentração de Células (kg/m³)
Cs=(Ks*D)/(μmax-kd-D) #Eq.(18)
# Concentração de Células (kg/m³)
Cx=Yxs*(Cse-Cs) #Eq.(19)
# Concentração de Produto (kg/m³)
Cp=((D*Cpe)+(α*μmax*Cx)+(β*Cx))/(D+Kdp) #Eq.(17)

# --------Processo com reciclo externo----------------

# Concentração de Células (kg/m³)
Cs_ext=Ks*(D*(1+r*(1-c))+kd)/(μmax-kd-D*(1+r*(1-c))) #Eq.(35)
# Concentração de Células (kg/m³)
Cx_ext=Yxs*(Cse-Cs_ext)*D/((μmax*Cs_ext/(Ks+Cs_ext))-kd) #Eq.(36)
#Cx_ext=(Cse-Cs_ext)*(Yxs/(1+r*(1-c)))
# Concentração de Produto (kg/m³)
Cp_ext=((D*Cpe)+(α*μmax*Cx_ext)+(β*Cx_ext))/(D+Kdp) #Eq.(34)

#Produtividade
P = Cx*D
P_ext = Cx_ext*D

# Ignorar Concentrações Indefinidas

for i in range(len(D)):
    if D[i]>=μmax or Cs[i]<0 or Cs[i]>=Cse:
      Cs[i]=Cse
      Cp[i]=Cpe
      Cx[i]=Cxe
      P[i] = 0
    if Cx_ext[i]<0 or Cs_ext[i]<0 or Cs_ext[i]>=Cse:
      Cs_ext[i]=Cse
      Cp_ext[i]=Cpe
      Cx_ext[i]=Cxe
      P_ext[i] = 0

warnings.filterwarnings("ignore")

# Construção do Gráfico

plt.figure(figsize=(10,6), facecolor="white", edgecolor="black")

plt.plot(D, Cx, label = "Células", color="black", linewidth=2, linestyle="-")
plt.plot(D, P, label = "Produtiv", color="black", linewidth=2, linestyle="--")

plt.plot(D, Cx_ext, label = "Reciclo ext - Células", color="green", linewidth=2, linestyle="-")
plt.plot(D, P_ext, label = "Reciclo ext - Produtiv", color="green", linewidth=2, linestyle="--")

plt.xlabel("Taxa de Diluição (1/h)", fontdict={'fontsize':12})
plt.ylabel("Concentração (kg/m³)", fontdict={'fontsize':12})

plt.ylim([-5, 30])
plt.xlim([0, 1])

plt.legend(fontsize=12)

plt.show()
```
