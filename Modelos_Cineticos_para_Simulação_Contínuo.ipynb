#1. Contínuo sem Reciclo

Contínuo.png

Um processo contínuo sem reciclo é carecterizado pela entrada e saída de meio, de maneira que o volume presente no reator não varie, sem realimentação de células no processo.

Portanto, no estado estacionário a variação mássica que representa o acúmulo é zero:


>$\frac{\mathrm{dm} }{\mathrm{d} t} = 0$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀Eq. (1)

##1.1. Balanço de Massa Geral

Para Contínuo, não há variação de volume, portanto, a vazão volumétrica de entrada deve ser igual a de saída, como apresentado pela Eq.(2):

>$Q_{e} = Q$  ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (2)

##1.2. Balanço de Massa Celular

Acúmulo = Entra - Sai + Geração - Consumo

⠀

Para Contínuo o balanço geral de células é simplificado para:


> 0 $=$ Entra - Sai + Geração - Morte

⠀⠀⠀⠀⠀⠀⠀⠀⠀$\therefore$

> $0= F_{xe} - F_{x} + R_{x} - R_{d}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (3)



---

Analogamente ao semicontínuo, define-se a **taxa de diluição** como $D = \frac{Q_{e}}{V} = \frac{Q}{V} $


---



Substituindo massa $m_{x}$ pela concentração celular $C_{x}$ utilizando a relação $C_{x}=\frac{m_{x}}{V}$, a vazão mássica $F_{xe}$ passa a ser descrita por $Q_{e}C_{xe}(\frac{m^3}{h}\frac{kg}{m^3})$, a geração celular passa a ser descrita por $r_{x}=\frac{R_{x}}{V}$ e a morte celular por $r_{d}=\frac{R_{d}}{V}$, dividindo por $V$ e reorganizando a equação em função de $D$:

$0 = Q_{e}C_{xe} - QC_{x} + R_{x} - R_{d}$
⠀

$\frac{0}{V} = \frac{QC_{xe}}{V} - \frac{QC_{x}}{V} + \frac{R_{x} - R_{d}}{V}$ ⠀

⠀⠀

>$D(C_{x} - C_{xe}) = r_{x} - r_{d}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (4)⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀


Analogamente a Batelada Simples e Alimentada, as taxas de crescimento e morte celular são descritas a seguir uitlizando a relação para reações de primeira ordem:

$r_{x}=\frac{\mathrm{d}[C] }{\mathrm{d} t} = k_{1}[C]^{1}$

$r_{d}=-\frac{\mathrm{d}[C] }{\mathrm{d} t} = k_{2}[C]^{1}$

---

Na qual: $[C]^{1} = C_{x}$ ; $k_{1}=\mu$ (taxa de crescimento celular) e $k_{2}=k_{d}$ (taxa de morte celular)

---

Então, a relação dada pela Eq.(4) passa a ser descrita pela Eq.(5) e a taxa de diluição é representada pela Eq.(6):

$D(C_{x} - C_{xe}) = (\mu - k_{d})C_{x}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (5)

⠀⠀

>$D = \frac{(\mu - k_{d})C_{x}}{C_{x} - C_{xe}}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (6)

No sistema em contínuo sem reciclo de células, considera-se a entrada de células $C_{xe} = 0$ e, para  $k_{d}<<\mu$, $D$ é representado pela Eq.(7)

$D = \frac{(\mu - k_{d})C_{x}}{C_{x}}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀

$D = (\mu - k_{d})$

>$D = \mu$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀Eq. (7)

##1.3. Balanço de Massa de Substrato

Acúmulo = Entra - Sai + Geração - Consumo

⠀

Para Contínuo o balanço geral de células é simplificado para:


> 0 $=$ Entra $-$ Sai $-$ Consumo

⠀⠀⠀⠀⠀⠀⠀⠀⠀$\therefore$

> $0= F_{se} - F_{s} - R_{s}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (8)

Utilizando a as relações de consumo de substrato $r_{s}=\frac{R_{s}}{V}$, de concentração de substrato $C_{s}=\frac{m_{s}}{V}$,  a vazão mássica $F_{s}$ passa a ser descrita por $QC_{s}(\frac{m^3}{h}\frac{kg}{m^3})$ e de taxa de diluição $D = \frac{Q}{V}$ reorganiza-se a Eq.(8) na Eq.(9):

0 = $Q_{e}C_{se} - QC_{s} -  R_{s} $

$\frac{Q_{e}C_{se}}{V} = \frac{QC_{s}}{V} + \frac{R_{s}}{V}$

$DC_{se}  = DC_{s} +  r_{s} $

⠀⠀

>$D(C_{se}-C_{s})  = r_{s} $⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (9)

Para relacionar o consumo de substrato com a formação celular, é necessário definir o **rendimento de crescimento celular** $Y_{\frac{x}{s}}$ (a quantidade de célula que é formada por substrato consumido):


$Y_{\frac{x}{s}} = \frac{r_{x}}{r_{s}}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀

⠀⠀

>$r_{s} = \frac{r_{x}}{Y_{\frac{x}{s}}}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀Eq. (10)

Unindo-se a Eq. (9) e (10) e utilizando a relação de $r_{x}=(\mu - k_{d})C_{x}$, a taxa de diluição $D$ é representada pela Eq.(11) quando $k_{d}<<\mu$ :

$D(C_{se}-C_{s})  = \frac{\mu C_{x}}{Y_{\frac{x}{s}}} $

⠀⠀
>$D  = \frac{(\mu - k_{d})C_{x}}{Y_{\frac{x}{s}}(C_{se}-C_{s})} $⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (11)



##1.4. Balanço de Massa de Produto

Acúmulo = Entra - Sai + Geração - Consumo

⠀

Para Contínuo o balanço geral de produto é simplificado para:


> 0 $=$ Entra - Sai + Gerado - Degradado

⠀⠀⠀⠀⠀⠀⠀⠀⠀$\therefore$

> $0 = F_{pe} - F_{p} + R_{p} - R_{dp}$⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (13)

Utilizando a as relações de consumo de substrato $r_{p}=\frac{R_{p}}{V}$, de concentração de produto $C_{p}=\frac{m_{p}}{V}$ e de taxa de diluição $D = \frac{Q}{V}$ reorganiza-se a Eq.(13) na Eq.(14):

$\frac{Qe}{V}C_{pe} = \frac{Q}{V}C_{p} + \frac{R_{dp}}{V} - \frac{R_{p}}{V}$


$D(C_{pe}-C_{p})  =  r_{dp} - r_{p}$

⠀⠀

>$D = \frac{r_{dp} - r_{p}}{C_{pe}-C_{p}} $⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (14)

Para relacionar a geração de produto com a formação celular, é necessário definir seu tipo de geração, que afetará o valor da concentração de produto obtida a partir da taxa de diluição da Eq.(14), na qual $\alpha$ e $\beta$ são constantes e $r_{dp} = k_{dp}C_{p}$:
⠀

⠀

>**Associado ao crescimento:** ⠀$D = \frac{k_{dp}C_{p} - \alpha \mu C_{x}}{C_{pe}-C_{p}} $

>$C_{p} = \frac{DC_{pe} + \alpha \mu C_{x}}{D+k_{dp}}$  ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ ⠀Eq. (15)

⠀

>**Não associado ao crescimento:**  $D = \frac{k_{dp}C_{p} - \beta C_{x}}{C_{pe}-C_{p}} $

>$C_{p} = \frac{DC_{pe} + \beta C_{x}}{D+k_{dp}}$  ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ ⠀Eq. (16)

⠀

>**Parcialmente associado ao crescimento:** ⠀$D = \frac{k_{dp}C_{p} - \alpha \mu C_{x}-\beta C_{x}}{C_{pe}-C_{p}} $

>$C_{p} = \frac{DC_{pe} + \alpha \mu C_{x} + \beta C_{x}}{D+k_{dp}}$  ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀Eq. (17)



## 1.5. Modelo cinético não estruturado para simulação **sem inibição**

Utilizando a Eq.(7) como base para $C_{s}$, Eq.(11) como base para $C_{x}$ e Eq.(17) para $C_{p}$, substitui-se $\mu$ constante por modelos cinéticos de $\mu$ variável. Para o modo contínuo vamos utilizar o modelo proposto por Monod (1949), entretanto quaisquer modelos já apresentados nos modos de operação de Batelada Simples e Alimentada podem ser substituidos.

### 1.5.1. (Monod, 1949)

$\mu = \frac{\mu_{max}C_{s} }{K_{s}+C_{s}}$

Utilizando o balanço celular para $C_{s}$:

$D = (\frac{\mu_{max}C_{s} }{K_{s}+C_{s}} - k_{d})$

Colocando $C_{s}$ em evidência:

>$ C_{s}= \frac{K_{s}D}{(\mu_{max}-k_{d}-D)}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (18)

Utilizando o balanço de substrato para $C_{x}$:

$D  = \frac{(\mu - k_{d})C_{x}}{Y_{\frac{x}{s}}(C_{se}-C_{s})} $

$D  = \frac{DC_{x}}{Y_{\frac{x}{s}}(C_{se}-C_{s})} $

Colocando $C_{x}$ em evidência:

$ C_{x}= \frac{{Y_{\frac{x}{s}}(C_{se}-C_{s})D}}{(\frac{\mu_{max}C_{s} }{K_{s}+C_{s}}- k_{d})}$

$ C_{x}= \frac{{Y_{\frac{x}{s}}(C_{se}-C_{s})D}}{D}$

>$ C_{x}= {Y_{\frac{x}{s}}(C_{se}-C_{s})}$ ⠀⠀⠀⠀ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (19)

Assim, unindo-se as Eq.(18) e (19):

>$ C_{x}= {Y_{\frac{x}{s}}(C_{se}-\frac{K_{s}D}{(\mu_{max}-k_{d}-D)})}$ ⠀⠀⠀⠀ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (20)

##1.6. Taxa de diluição ótima ($D_{ótimo}$)


Partindo da definição de **produtividade volumética de células** (g/L/h), dado a seguir, calcula-se a diluição na qual a produtividade é máxima, ou seja, sua derivada é zero:

$P = \frac{Q_{e}C_{x}}{V} = DC_{x}$

⠀⠀

---


Na qual $D=D_{ótimo}$ quando $\frac{\mathrm{dP} }{\mathrm{d} D} = 0$


---



$\frac{\mathrm{dDC_{x}} }{\mathrm{d} D} = 0$

$\frac{\mathrm{d}}{\mathrm{dD}}{Y_{\frac{x}{s}}(C_{se}-C_{s})}D = 0$

$\frac{\mathrm{d}}{\mathrm{dD}}{Y_{\frac{x}{s}}(C_{se}-\frac{K_{s}D}{(\mu_{max}-k_{d}-D)})}D = 0$

Derivando a equação anterior:


```python
import sympy as sp
from sympy import diff

D, Yxs, Cse, Ks, μmax, kd = sp.symbols('D Yxs Cse Ks μmax kd')
func=Yxs*(Cse-(Ks*D)/(μmax-kd-D))*D

diff(func,D)
```

Obtém-se o $D_{ótimo}$ igualando a equação acima a zero e deixando $D$ em evidência:
⠀
>$D_{ótimo} = -\frac{\sqrt{(k_{d} - \mu_{max})^2 K_{s}(C_{se} + K_{s})}}{K_{s}+C_{se}}-k_{d}+\mu_{max}$ ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀ Eq. (21)

Reorganizando:

>$D_{ótimo} = (\mu_{\max} - k_d) \left( 1- \sqrt{\frac{K_s}{K_s + C_{se}}} \right)$

---

Definindo a **tempo de residência** como $\tau = \frac{1}{D}$


---

>$\tau_{ótimo} = \frac{1}{D_{ótimo}}$

Teoricamente, para a reação em contínuo entrar em estado estacionário, em geral demora 3x o tempo de residência
